```sql
DROP TABLE IF EXISTS Content;

CREATE TABLE Content(
    code VARCHAR(2),
    name VARCHAR(100),
    cases INT,
    deaths INT,
    averagetemp INT,
    day DATE NOT NULL
) 
AS SELECT code, name, cases, deaths, averagetemp, day
FROM CSVREAD('https://raw.githubusercontent.com/MatheusBulhoes/MC536-Trabalho/main/final/content.csv')
```

```sql
DROP VIEW IF EXISTS Covid_Percent;

CREATE VIEW Covid_Percent AS
SELECT
C.code,
C.name,
C.cases as Cases, 
COALESCE((C.cases - CL.cases)/ NULLIF(CAST(CL.cases AS FLOAT), 0), 0) + 0.0 AS CasesVariation,
C.averagetemp,
C.day
FROM Content AS C, Content AS CL
WHERE CL.day = DATEADD(day, -1, C.day) AND C.code = CL.code
ORDER BY C.code, C.day;

SELECT * FROM Covid_Percent
```
